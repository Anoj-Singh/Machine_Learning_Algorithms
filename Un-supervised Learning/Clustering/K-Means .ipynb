```python
#importing libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#importing dataset

df=pd.read_csv(r"D:/csv files/Mall_Customers_clu.csv")
x=df.iloc[:,[3,4]].values
```

```python
#using elbow method to find the optimal number of clusters

from sklearn.cluster import KMeans

wcss=[]
for i in range(1,11):
    model=KMeans(n_clusters=i,init='k-means++',random_state=42)
    model.fit(x)
    wcss.append(model.inertia_)
plt.plot(range(1,11),wcss)
plt.show()
```

```python
#training the model

model=KMeans(n_clusters=5,init='k-means++',random_state=42)
y_pred=model.fit_predict(x)
```

```python
#visualizing the clusters

plt.scatter(x[y_pred==0,0],x[y_pred==0,1],label='Clustrer 1')
plt.scatter(x[y_pred==1,0],x[y_pred==1,1],label='Clustrer 2')
plt.scatter(x[y_pred==2,0],x[y_pred==2,1],label='Clustrer 3')
plt.scatter(x[y_pred==3,0],x[y_pred==3,1],label='Clustrer 4')
plt.scatter(x[y_pred==4,0],x[y_pred==4,1],label='Clustrer 5')
plt.scatter(model.cluster_centers_[:,0],model.cluster_centers_[:,1],s=200,label='Centtroids',marker='*',c='r')
plt.legend()
plt.show()
```

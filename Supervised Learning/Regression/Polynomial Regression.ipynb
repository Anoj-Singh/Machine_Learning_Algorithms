```python
#importing libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#importing dataset

df=pd.read_csv("D://Position_Salaries.csv")
x=df.iloc[:,1:-1].values
y=df.iloc[:,-1].values
```

```python
plt.scatter(x,y)
plt.show()
```

```python
#training linear regression model

from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x,y)
```

```python
#visualizing linear regression results

plt.scatter(x,y,color='r')
plt.plot(x,lr.predict(x))
plt.title("PREDICTING SALARIES CORRESPONDING LEVEL")
plt.xlabel("LEVEL")
plt.ylabel("SALARY")
plt.show()
```

```python
#training polynomial model

from sklearn.preprocessing import PolynomialFeatures
poly_reg=PolynomialFeatures(degree=4)
x_polly=poly_reg.fit_transform(x)
lin_reg2=LinearRegression()
lin_reg2.fit(x_polly,y)
```

```python
#visualizing polynomial dataset results

plt.scatter(x,y,color='r')
plt.plot(x,lin_reg2.predict(x_polly),color='b')
plt.title("PREDICTING SALARIES CORRESPONDING LEVEL")
plt.xlabel("LEVEL")
plt.ylabel("SALARY")
plt.show()
```

```python
#linear regression prediction

lr.predict([[6.5]])
```

```python
#polynomail regression results

lin_reg2.predict(poly_reg.fit_transform([[6.5]]))
```
